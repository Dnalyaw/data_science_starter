## Segment 1 - Simple linear regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pylab import rcParams

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 10,8
```

```python
rooms = 2*np.random.rand(100, 1) + 3 #100 by 1 2d array that is multiplied by 2 and adds 3, so in the range of 3-5
rooms
```

```python
rooms[0:5] + 2 #adds to all elements
```

```python
rooms[0:5] + np.array([1, 2]) #but an array would add the anumber to each in the array, if added to 2d array
```

```python
print(np.random.randn(100, 1))
plt.plot(np.random.randn(100, 1))#means mean of 0 and variance of 1
```

```python
homePrice = 265 + 6*rooms + abs(np.random.randn(100, 1))#each item is multiplied by 6, added to 100 by 1 2d array creating 100*1
print(homePrice[1:10])
homePrice.shape
```

```python
plt.plot(rooms, homePrice, 'r^')
plt.xlabel("Number of Rooms, 2019 Average")
plt.ylabel("2019 Average Home, in 1000s of USD")
#because in the equation above the rooms were directly proportional to the random numbers 
```

```python
X = rooms 
y = homePrice

LinReg = LinearRegression()
LinReg.fit(X, y)
print(LinReg.intercept_, LinReg.coef_) #using the ys and xs, we created a y = mx +b line for the estimation scatterplot, b = intercept and m = coef, or slope 
```

*Simple Algebra*
- y = mx + b
- b = intercept = 265.7

*Estimated Coefficients*
- LinReg.coef_ = [5.99] Estimated coefficients for the terms in the linear regression problem.

```python
LinReg.score(X, y) #r^2, where r is the relationship between the data and the linear regression line, the number of rooms explains 97% of the data, something else is explaining the other 3%
```
