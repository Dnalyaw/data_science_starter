### The simplest time series plot

```python
address = "C:/Users/wayla/Python Data Analysis Exercise Files/Ex_Files_Python_Data_Science_EssT_Pt_1/Exercise Files/Data/Superstore-Sales.csv"

df = pd.read_csv(address, index_col='Order Date', encoding='cp1252', parse_dates=True)
df.head()
```

```python
df2 = df.sample(n=100, random_state=25, axis=0)#takes 100 random rows from the list, jumbles them up

plt.xlabel('Order Date')
plt.ylabel('Order Quantity')
plt.title('Superstore Sales')
df2['Order Quantity'].plot()
plt.grid()
plt.show()
```
