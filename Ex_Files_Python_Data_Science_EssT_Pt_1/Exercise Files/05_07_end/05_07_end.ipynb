### Normalizing and transforming features with MinMaxScalar() and fit_transform()

```python
address = "C:/Users/wayla/Python Data Analysis Exercise Files/Ex_Files_Python_Data_Science_EssT_Pt_1/Exercise Files/Data/mtcars.csv"


cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
```

```python
mpg = cars.mpg
plt.plot(mpg)
```

```python
cars["mpg"]
```

```python
cars[['mpg']].describe()
```

```python
print(mpg.values)#.values creates a list with many items
mpg.values.reshape(-1,1)#states we are unaware of the amount of rows but reshape with one column, so creates 2d array
```

```python
mpg_matrix = mpg.values.reshape(-1,1)

scaled = preprocessing.MinMaxScaler()#this is the recipe to create the same graph but from 0-1

scaled_mpg = scaled.fit_transform(mpg_matrix)#applying it to the data
plt.plot(scaled_mpg)#highest value is now one, lowest is 0
```

```python
scaled = preprocessing.MinMaxScaler(feature_range=(0,10))#it doesn't have to be 0-1, it can be whatever

scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg)
```

### Using scale() to scale your features

```python
standardized_mpg = scale(mpg, axis=0, with_mean=False, with_std=False)#no mean value or std
plt.plot(standardized_mpg)#notice how the values are the same
```

```python
standardized_mpg = scale(mpg)#this actually standardizes it
plt.plot(standardized_mpg)#standardized: mean is set to 0, graph cancels itself out
```
