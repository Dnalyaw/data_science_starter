```python
xdata = np.array([0,1,4,5,7,8], 'd')
ydata = np.array([10,5,2,7,7.5,10], 'd')

pp.plot(xdata, ydata, '--o') #plotting x and y points in a graph
```

```python
# interpolate x/y data with missing values to continuous x values

xnew = np.linspace(0, 8, 9)#array of 0-8
ynew = np.interp(xnew, xdata, ydata)#interp interprets what the y data could be using the previous points given to fill in what is missing

ynew
```

```python
pp.plot(xdata, ydata, '--o', ms=10)
pp.plot(xnew, ynew, 's')#orange dots show what points were interpreted to now account for all values
```

```python
# interpolate x/y data with missing values to denser, continuous x values

xnew = np.linspace(0, 8, 30) #0-8 cut into 30 pieces
ynew = np.interp(xnew, xdata, ydata)

pp.plot(xdata, ydata, '--o', ms=10)
pp.plot(xnew, ynew, 's')
```

```python
pasadena = getweather.getyear('PASADENA', ['TMIN', 'TMAX'], 2001) #now that we have replaced nans with means, change database
```

```python
# build a Boolean mask of "good" (non-NaN) TMIN values;
# interpolate "good" days/TMIN to full range of days 

good = ~np.isnan(pasadena['TMIN'])
x = np.arange(0, 365)

good
```

```python
# fill NaNs in any array by interpolation

def fillnans(array):
    good = ~np.isnan(array)
    x = np.arange(len(array))

    return np.interp(x, x[good], array[good])
```

```python
pp.plot(fillnans(pasadena['TMIN']))
pp.plot(fillnans(pasadena['TMAX']))
```
