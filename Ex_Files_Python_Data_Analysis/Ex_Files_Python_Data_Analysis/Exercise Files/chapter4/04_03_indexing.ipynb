```python
pp.imshow(monalisa[400:800,200:600,0:3]) # (section of pixels of the height 400-800), (section of pixels of width 200-600), all the colors
```

```python
pp.imshow(monalisa[400:800,:,:])
```

```python
pp.imshow(monalisa[400:800,...]) #three dots means keep everything, just like the one above
```

```python
pp.imshow(monalisa[::20,::20,:])#everything but in steps of 20
```

```python
pp.imshow(monalisa[::-20,::20,:]) #counts from the back rows, so they become the front rows turning the photo upside down
```
