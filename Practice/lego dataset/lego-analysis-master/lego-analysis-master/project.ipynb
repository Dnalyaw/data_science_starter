```python
# Use this cell to begin your analyses, and add as many cells as you would like!
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
parent_theme = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
```

```python
merged = df.merge(parent_theme, left_on='parent_theme', right_on="name")#merge the parent theme to df
merged = merged.dropna(subset=["set_num"]) #left on = col of df looking to match, right on = col of parent theme looking to match
```

```python
#column name in both dataframes so delete one
merged = merged.drop('name_y', axis=1)
```

```python
merged = merged.rename(columns={'name_x':'name'})
merged
```

```python
licensed = merged.loc[merged["is_licensed"]]
```

```python
licensed
```

```python
licensed[licensed['parent_theme'] == "Star Wars"].shape[0]
```

```python
licensed.shape[0]
```

```python
licensed[licensed['parent_theme'] == "Star Wars"].shape[0] / licensed.shape[0]
```

```python
new_df = licensed.groupby(['year','parent_theme']).set_num.count().sort_values(ascending=False).reset_index()
```

```python
new_df = new_df.sort_values('year')
new_df = new_df.rename(columns={"set_num":'count'})
new_df
```

```python
new_df = new_df.sort_values('count', ascending=False) #so we only want the rows where the count is the greatest number of the year
new_df#now we have it filtered from highest to lowest count
```

```python
new_df.drop_duplicates(["year"]).sort_values('year')#by dropping the duplicate years, it keeps the first iteration (highest count) but takes away the rest, leaving the greatest for each year
```
