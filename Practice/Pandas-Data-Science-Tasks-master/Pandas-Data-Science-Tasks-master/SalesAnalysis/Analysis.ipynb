```python
files = [file for file in os.listdir('Sales_Data')]
year_df = pd.DataFrame()

for file in files:
    df = pd.read_csv("Sales_Data/"+file)
    year_df = pd.concat([year_df, df])#add all 12 months together 
```

```python
year_df = year_df.dropna().reset_index(drop=True)#some NANs; (how="any") drops a row if it is has a singular NAN
```

```python
year_df = year_df[year_df["Order ID"] != "Order ID"]#since i added all the files together, the top columns transferred to, so this states to get rid of the other ones
```

```python
year_df["Month"] = year_df["Order Date"].str[0:2]#month colu,n
```

```python
year_df.iloc[516] #example
```

```python
year_df = year_df.apply(pd.to_numeric, errors='ignore') #change datatypes which are strings to ints/floats
```

```python
year_df.dtypes
```

```python
year_df["Total"] = year_df["Quantity Ordered"] * year_df["Price Each"] #now we can do math since these are numbers and not strings
```

```python
results = year_df.groupby('Month').Total.sum()#shows that December has the highest sales making $4613443.34 in sales
results
```

```python
import matplotlib.pyplot as plt

months = range(1, 13)

plt.bar(months, results)#x, y
plt.xticks(months)
plt.ylabel("Sales (in Millions of USD)")
plt.xlabel('Month Number')
plt.show()
```

```python
def cityName(address):#using function to use apply method
    i = address.index(',')
    return address[i+2:-10]
```

```python
boughtTogether = year_df[year_df['Order ID'].duplicated(keep=False)]#.duplicated() gives trues and falses for if the second occurence is a duplicate, but keep=false puts both of the matching Orer IDs to true
boughtTogether #should only print out the ones with multiple of the same ID numbers
```

```python
boughtTogether['Grouped'] = boughtTogether.groupby("Order ID")['Product'].transform(lambda x: ','.join(x))#by grouping together, then adding a line with both items in it, we will get to see what items are bought together often
boughtTogether #now one row with the names of both
```

```python
boughtTogether = boughtTogether[['Order ID', 'Grouped']].drop_duplicates()#getting rid of the duplicates
boughtTogether #now will only show the rows with order id and items with the matching duplicate dropped
```

```python
boughtTogether.groupby('Grouped').Grouped.count().sort_values(ascending=False).head(10) #problem with this is that it could be in a different order but the same values
```

```python
from itertools import combinations
from collections import Counter

count = Counter()#so how can we tell the difference between "item1,item2" and "item2,item1"

for row in boughtTogether['Grouped']:
    row_list = row.split(',')#["item 1", "item 2"]
    count.update(Counter(combinations(row_list, 2)))#Counting "unique pairs" of numbers into a python dictionary?
count
```

```python
count.most_common(10)
```
